```python


def leading_trailing(grammar):
    leading = {}
    trailing = {}

    for nt in grammar:
        leading[nt] = set()
        trailing[nt] = set()

        for prod in grammar[nt]:
            # LEADING → first symbol
            if prod[0].islower():
                leading[nt].add(prod[0])

            # TRAILING → last symbol
            if prod[-1].islower():
                trailing[nt].add(prod[-1])

    return leading, trailing


# Example grammar
grammar = {
    'E': ['E+T', 'T'],
    'T': ['T*F', 'F'],
    'F': ['(E)', 'i']
}

l, t = leading_trailing(grammar)

print("LEADING:", l)
print("TRAILING:", t)
```
